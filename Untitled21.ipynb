## Penjelasan Struktur Dataset

Dataset Titanic memiliki 891 baris dan 12 kolom.

Dataset ini memiliki beberapa tipe data, yaitu:

- int64: PassengerId, Survived, Pclass, SibSp, Parch
- float64: Age, Fare
- object: Name, Sex, Ticket, Cabin, Embarked

Penjelasan:
- int64 adalah tipe data angka bulat
- float64 adalah tipe data angka desimal
- object adalah tipe data teks
